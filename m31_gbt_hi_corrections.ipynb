### Post-hoc corrections for lingering baseline issues

```python
%matplotlib inline
```

```python
from spectral_cube import SpectralCube
```

```python
%cd /reduction/erickoch/LGLBS/hi_feathering
```

```python
cube = SpectralCube.read("M31_GBT.FITS")
```

```python
# Residual ripples are mostly beyond -600
```

```python
cube_clipped = cube.sigma_clip_spectrally(3)
```

```python
median_clipped = cube_clipped.spectral_slab(-300*u.km/u.s, 0*u.km/u.s).median(axis=0)

median_clipped[median_clipped.value >= 0.0] = 0.0
```

```python
plt.imshow(median_clipped.value, origin='lower')
plt.colorbar()
plt.contour(median_clipped.value < -0, colors='g')
```

```python
_ = plt.hist(median_clipped.value[np.isfinite(median_clipped)], bins=100)
```

```python
mad_clipped = cube_clipped.spectral_slab(-500*u.km/u.s, 0*u.km/u.s).mad_std(axis=0)
```

```python
plt.imshow(mad_clipped.value, origin='lower')
plt.colorbar()
plt.contour(mad_clipped.value < 0.115, colors='g')
```

```python
_ = plt.hist(mad_clipped.value[np.isfinite(mad_clipped)], bins=100)
```

```python
np.median(mad_clipped.value[np.isfinite(mad_clipped)])
```

### Correct cube with oversubtracted regions

```python
cube_corrected = cube - median_clipped
```

```python
cube_corrected.write("M31_GBT_corrected.fits", overwrite=True)
```
